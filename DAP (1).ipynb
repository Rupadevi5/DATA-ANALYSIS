```python

import numpy as np
arr=np.array([1,2,3,4,5])         #creating array
print(arr)

zeros_arr=np.zeros((3,3),dtype=int)       #dtype for type conversion
print(zeros_arr)

ones_arr=np.ones((2,2),dtype=int)
print(ones_arr)

arange_arr=np.arange(10)
print(arange_arr)

reshaped_arr=arr.reshape(5,1)
print(reshaped_arr)

sliced_arr=arr[2:4]          #particular accessing of values
print(sliced_arr)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
b=pd.read_csv("/content/temp.csv")
print(b)

avg_temp=b['temperature'].mean()
high_temp=b['temperature'].max()
low_temp=b['temperature'].min()
threshold=30
days_above_threshold=b[b['temperature']>threshold].shape[0]

print("Average temperature:",avg_temp)
print("Highest temperature:",high_temp)
print("Lowest temperature:",low_temp)
print("Number of days temperature exceeded",threshold,"degree celsius:",days_above_threshold)

b.plot(kind='line',color='red')
plt.title("Temperature for 50 days")
plt.xlabel("Days")
plt.ylabel("Temperature")
```
